```python
# group total fights per pokemon
total_first_pokemons = combats.groupby('First_pokemon').count()
total_second_pokemons = combats.groupby('Second_pokemon').count()
total_winners = combats.groupby('Winner').count()
```

```python
print("total first pokemons: " + str(total_first_pokemons.shape))
print("total second pokemons: " + str(total_second_pokemons.shape))
print("total pokemons that won: " + str(total_winners.shape))
```

The pokemon that never won

```python
loser = np.setdiff1d(total_first_pokemons.index.values, total_winners.index.values)
#print(poke_loser[0])
poke_loser = pokemons.loc[pokemons['#'] == loser[0]]
poke_loser
```

List of Pokemons with no recorded fights

```python
p_nofights = np.setdiff1d(pokemons['#'], total_winners.index.values)
pokemons_nofight = pokemons.loc[pokemons['#'].isin(p_nofights)]
pokemons_nofight
```

```python
first_wins = combats.loc[combats['First_pokemon'] == combats['Winner']]
print(len(first_wins))
first_win_percentage = len(first_wins) / len(combats)
print("win percentage attacking first Pokemon: " + str(first_win_percentage))
```

```python
second_wins = combats.loc[combats['Second_pokemon'] == combats['Winner']]
print(len(second_wins))
second_win_percentage = len(second_wins) / len(combats)
print("win percentage attacking second Pokemon: " + str(second_win_percentage))
```

```python
total_winners['Total_fights'] = total_first_pokemons.Second_pokemon + total_second_pokemons.First_pokemon
total_winners['Win_percentage'] = total_winners.First_pokemon / total_winners.Total_fights
```

```python
total_winners.head()
```

```python
pokemon_data = pokemons.merge(total_winners, right_index=True, left_on='#')
```
